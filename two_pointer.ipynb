### Problem 1: TWO SUM II

You’re given a sorted array of integers called numbers and an integer target. Your job is to find two numbers in the array that add up exactly to target. You return their positions, and on LeetCode this problem specifically wants 1-indexed positions. You’re guaranteed exactly one solution, and you can’t use the same element twice.

```python
def two_sum_ii(numbers, target):
    left = 0
    right = len(numbers) - 1

    while left < right:
        s = numbers[left] + numbers[right]

        if s == target:
            return [left + 1, right + 1]  # LeetCode wants 1-indexed
        elif s < target:
            left += 1
        else:
            right -= 1

    return None
```

```python
print(two_sum_ii([2, 7, 11, 15], 9))      # [1, 2]
print(two_sum_ii([2, 3, 4], 6))           # [1, 3]
print(two_sum_ii([-1, 0], -1))            # [1, 2]
```

### Problem 2: Valid Palindrome

You’re given a string s. You need to determine if it’s a palindrome, meaning it reads the same forward and backward, but with two rules: ignore punctuation/spaces, and ignore letter casing. Only alphanumeric characters count.

```python
def is_palindrome(s):
    left = 0
    right = len(s) - 1

    while left < right:
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1

        if s[left].lower() != s[right].lower():
            return False

        left += 1
        right -= 1

    return True
```

```python
print(is_palindrome("A man, a plan, a canal: Panama"))  # True
print(is_palindrome("race a car"))                      # False
print(is_palindrome(" "))                               # True
print(is_palindrome("0P"))                              # False
```

### Problem 3: Container with Most Water

You’re given an array height where each value represents a vertical line on a chart. The lines at positions i and j form a container, and the amount of water it can hold is the distance between them times the smaller of the two heights. Your goal is to find the maximum possible area.

```python
def max_area(height):
    left = 0
    right = len(height) - 1
    best = 0

    while left < right:
        h = min(height[left], height[right])
        w = right - left
        best = max(best, h * w)

        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return best
```

```python
print(max_area([1,8,6,2,5,4,8,3,7]))  # 49
print(max_area([1,1]))                # 1
```

### Problem 4: Remove Duplicates From Sorted Array

You’re given a sorted array nums, and you need to remove duplicates in-place so that each unique value appears only once. You return k, the number of unique elements, and the first k positions of nums should contain the unique values. The order must stay the same.

```python
def remove_duplicates(nums):
    if not nums:
        return 0

    write = 1

    for read in range(1, len(nums)):
        if nums[read] != nums[read - 1]:
            nums[write] = nums[read]
            write += 1

    return write
```

```python
nums = [1,1,2]
k = remove_duplicates(nums)
print(k, nums[:k])  # 2 [1,2]

nums = [0,0,1,1,1,2,2,3,3,4]
k = remove_duplicates(nums)
print(k, nums[:k])  # 5 [0,1,2,3,4]
```
